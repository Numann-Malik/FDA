# Financial Data Analysis

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
pd.options.display.float_format = '{:.4f}'.format
```

```python
start = "2014-10-01"
end = "2023-04-30"
```

Ticker Symbols:
- BA: Boeing (US Stock)
- MSFT: Microsoft Corp (US Stock)
- ^DJI: Dow Jones Industrial Average (US Stock Index)
- EURUSD=X: Exchange Rate for Currency Pair EUR/USD (Forex)
- GC=F: Gold Price (Precious Metal / Commodity)
- BTC-USD: Bitcoin in USD (Cryptocurrency)

```python
symbol = ["BA", "MSFT", "^DJI", "EURUSD=X", "GC=F", "BTC-USD"]
```

```python
df = yf.download(symbol, start, end)
df
```

```python
df.to_csv("multi_assets.csv")
```

## Data Inspection

```python
df = pd.read_csv("multi_assets.csv", header = [0,1], index_col = 0, parse_dates =[0])
df
```

```python
# outer index level
df.Close
```

```python
# one column
df.Close.BA
```

```python
# one day
df.loc["2022-01-07"]
```

```python
# one month for one column
df.loc["2021-06":, ("Close", "BA")]
```

```python
# swap outer and inner index
df = df.swaplevel(axis = "columns").sort_index(axis = "columns")
df
```

```python
# select daily close prices only
close = df.Close.copy()
```

```python
close.describe()
```

```python
close.BA.dropna().plot(figsize = (15,8), fontsize = 13)
plt.legend(fontsize = 13)
plt.show()
```

```python
close.dropna().plot(figsize = (15,8), fontsize = 13)
plt.legend(fontsize = 13)
plt.show()
```

Note: Prices on a different scale cannot be compared.

## Normalizing Financial Time Series to a Base Value (100)

```python
norm = close.div(close.iloc[0]).mul(100)
norm
```

```python
norm.dropna().plot(figsize = (15,8), fontsize = 13, logy = True)
plt.legend(fontsize = 13)
plt.show()
```

Note: Normalized prices help to compare financial instruments but are limited when it comes to measuring/comparing performance in more detail.
